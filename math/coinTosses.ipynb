```python
import numpy as np
from math import ceil
import matplotlib.pyplot as plt
```

### Suppose we roll a dice n times, and we can either take home earnings or roll again. What do we do?

```python
N = 100
expectationsForN = [3.5]

for i in range(2, N):
    
    expectationFromStopping = expectationsForN[-1]
    expectationFromRollingOneMore = 0
    
    probabilityLeft = 1
    
    for j in range(ceil(expectationFromStopping), 7):
        expectationFromRollingOneMore += j*1/6
        probabilityLeft -= 1/6
    
    totalExpectation = expectationFromRollingOneMore + expectationFromStopping*probabilityLeft
    
    expectationsForN.append(totalExpectation)
    
    print(i, totalExpectation, expectationFromRollingOneMore, expectationFromStopping, probabilityLeft)
```

```python
plt.plot(expectationsForN)
```

### Suppose we have 2 dice with n face each, what is the expectation of the max of the two dice

```python
# Let X = max(a, b)
# Check by simulation
NUM_TRIALS = 10000
NUM_FACES = 1500
outputs = []

for n in range(1, NUM_FACES):
    a = np.random.randint(n, size=NUM_TRIALS)
    b = np.random.randint(n, size=NUM_TRIALS)
    X = np.maximum(a, b)
    expectedValue = X.mean()
    outputs.append(expectedValue)
```

```python
plt.plot(outputs)
```
